```python
cereals_df = pd.read_csv('./cereal.csv')
```

```python
cereals_df.head()
```

```python
sugary = cereals_df.sort_values(by = 'sugars', ascending = False)[:10]
```

```python
plt.figure(figsize = (15,5))
plt.bar(sugary ['name'], sugary['sugars'])
```

```python
plt.figure(figsize=(15,5))

plt.bar(sugary['name'], sugary['sugars'])

plt.title('Top ten sugariest cereals')
plt.xticks(rotation=10)

plt.xlabel('Cereal name')
plt.ylabel('Grams of sugar per bite');
```

```python
plt.figure(figsize=(12,2))
plt.hist(cereals_df['rating'], color = 'red')
plt.title("cereal ratings");
```

```python
plt.figure(figsize=(12,7))
plt.scatter(cereals_df['calories'], cereals_df['sugars'])
plt.title('sugar vs calories');
plt.xlabel('calories')
plt.ylabel('sugars');
```

```python
plt.figure(figsize=(16,19))
ax1 = plt.subplot(3,1,1)
ax1.set_title("Grams of protein per serving")
ax1.hist(cereals_df['protein'])
ax1.axvline(x= np.mean(cereals_df['protein']), color = 'yellow')


ax2 = plt.subplot(3,1,2)
ax2.set_title("Grams of Fat per serving")
ax2.hist(cereals_df['fat'],color = 'orange')
ax2.axvline(x= np.mean(cereals_df['fat']), color = 'green')

ax3 = plt.subplot(3,1,3)
ax3.set_title("Grams of Carbs per serving")
ax3.hist(cereals_df['carbo'],color = 'pink')
ax3.axvline(x= np.mean(cereals_df['carbo']), color = 'red');
```
